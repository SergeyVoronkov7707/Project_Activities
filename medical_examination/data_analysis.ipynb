```python
path = "./data"
print(os.listdir(path))

data = pd.read_csv(f"{path}/medical_examination.csv")
data.columns = data.columns.str.lower()

print("data.shape = {} rows, {} cols".format(*data.shape))
```

```python
categorical = data.dtypes[data.dtypes == "object"].index
numerical = list(set(data.columns) - set(categorical))
print("categorical columns data: len {}, {}".format(len(categorical), categorical))
print("len numerical columns data: {}".format(len(numerical)))
print("Nan data: {}".format(data.isna().sum()))
data.head(n=3)
```

```python
def q90(x):
    return x.quantile(0.99)
```
